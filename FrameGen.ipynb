```python
import os
import cv2

def get_video_fps(video_path):
    """Function to get the FPS of a video."""
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Could not open video: {video_path}")
        return None
    fps = cap.get(cv2.CAP_PROP_FPS)
    cap.release()
    return fps

def print_fps_for_all_videos_in_directory(directory_path):
    """Function to loop through all videos in a directory and print FPS."""
    if not os.path.exists(directory_path):
        print(f"Directory not found: {directory_path}")
        return
    
    # Loop through all video files in the directory
    for video_file in os.listdir(directory_path):
        video_path = os.path.join(directory_path, video_file)
        if video_file.endswith(('.mp4', '.avi', '.mov')):  # Check for valid video formats
            fps = get_video_fps(video_path)
            if fps:
                print(f"FPS of video {video_file}: {fps}")
            else:
                print(f"Skipping {video_file}, unable to read FPS.")
        else:
            print(f"Skipping non-video file {video_file}.")

# Example usage
directory_path = '../data/videos/real/celeb_real/'  # Replace with the actual path to your directory containing video files
print_fps_for_all_videos_in_directory(directory_path)
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNet
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from kerastuner.tuners import Hyperband
import PIL
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.losses import BinaryCrossentropy
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define paths
train_data_path = '../data/frames'  # Combined real and fake directory
image_size = (224, 224)
batch_size = 32

# Data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1.0/255,  # Normalize pixel values
    validation_split=0.2  # 80% training, 20% validation
)

# Training generator
train_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)

# Validation generator
val_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)
```

```python
labels = train_generator.classes  # Extract class labels from training generator
class_weights = compute_class_weight(
    class_weight='balanced', classes=np.unique(labels), y=labels
)
class_weights_dict = {i: weight for i, weight in enumerate(class_weights)}

def weighted_binary_crossentropy(class_weights):
    def loss(y_true, y_pred):
        weights = tf.where(tf.equal(y_true, 1), class_weights[1], class_weights[0])
        return BinaryCrossentropy()(y_true, y_pred) * weights
    return loss

print("Class Weights: ", class_weights_dict)
class_weights_dict = {key: float(value) for key, value in class_weights_dict.items()}

# Verify the dictionary
print(class_weights_dict)
```

```python
import keras_tuner as kt

# Define the model for tuning
def build_model(hp):
    base_model = MobileNet(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
    base_model.trainable = False

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(hp.Int('units', min_value=64, max_value=256, step=64), activation='relu'),
        layers.Dropout(hp.Float('dropout', min_value=0.2, max_value=0.5, step=0.1)),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[0.001, 0.0005, 0.0001])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Initialize the tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=10,
    factor=3,
    directory='tuner_dir',
    project_name='deepfake_detector'
)

# Perform hyperparameter search
tuner.search(train_generator, validation_data=val_generator, epochs=10)

# Get the best hyperparameters and rebuild the model
best_hps = tuner.get_best_hyperparameters(1)[0]
print(f"Best hyperparameters: {best_hps.values}")

# Build and train the best model
best_model = tuner.hypermodel.build(best_hps)
history = best_model.fit(train_generator, validation_data=val_generator, epochs=10)
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.callbacks import EarlyStopping
import keras_tuner as kt
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import RandomOverSampler
```

```python
# Define paths
train_data_path = '../data/frames'

# ImageDataGenerator with advanced augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    validation_split=0.2,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
X, y = next(train_generator)  # Get one batch from the generator (first 32 images)
batch_size = X.shape[0]  # Get batch size (this will be 32 in this case)

# Initialize arrays to collect the images and labels
X_all, y_all = X, y
num_images = 200  # Limit to 200 images

# Loop through the generator until we collect 200 images
for i in range((num_images // batch_size) - 1):  # Number of batches needed to get 200 images
    X_batch, y_batch = next(train_generator)
    X_all = np.vstack((X_all, X_batch))
    y_all = np.hstack((y_all, y_batch))

    # Print progress (how many images collected so far)
    print(f"Images collected: {(i + 1) * batch_size}/{num_images}")

# Confirm the total number of images collected
print(f"Total images collected: {X_all.shape[0]} / {num_images} images")

# Step 2: Apply oversampling using RandomOverSampler
oversampler = RandomOverSampler(sampling_strategy='minority')

# Reshape images for the oversampler (flatten them while keeping batch size as the first dimension)
X_all_reshaped = X_all.reshape(X_all.shape[0], -1)  # Flatten the images
X_resampled, y_resampled = oversampler.fit_resample(X_all_reshaped, y_all)

# Reshape the oversampled data back to the original image shape (after oversampling)
X_resampled = X_resampled.reshape(-1, 224, 224, 3)  # Using the image size and channels

# Step 3: Create a new data generator for the oversampled dataset
oversampled_datagen = ImageDataGenerator(rescale=1.0 / 255)
train_generator_resampled = oversampled_datagen.flow(
    X_resampled, y_resampled, batch_size=batch_size  # Use the original batch size
)
```

```python
# Model building function for Hyperparameter Tuning
def build_model(hp):
    # Load MobileNet as base model
    base_model = MobileNet(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
    base_model.trainable = False  # Freeze layers

    # Add custom layers
    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(hp.Int("units", min_value=64, max_value=256, step=64), activation='relu'),
        layers.Dropout(hp.Float("dropout", min_value=0.2, max_value=0.5, step=0.1)),
        layers.Dense(1, activation='sigmoid')  # Binary classification
    ])

    # Compile model
    model.compile(
        optimizer=Adam(learning_rate=hp.Choice("learning_rate", values=[1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```
